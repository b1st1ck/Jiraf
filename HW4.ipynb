```python
import matplotlib.pyplot as plt
import numpy as np
```

Задание 1

```python
p = round(np.pi,3)
x = np.linspace(-p,p,10000)
y = np.sin(1/x)
plt.plot(x,y,color='green')
ax = plt.gca()
ax.set_xlim(-p,p)
ax.set_ylim(-1,1)
ax.set_xticks([-p,-p/2,0,p/2,p])
ax.set_yticks([-1,0,1])
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

В моем примере это функция sin(1/x)
